```python
import random # import random package 

def hangman():
    # define the word list 
    words = ["WHDSPQZ", "XHO", "TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"]
    word = words[random.randint(0, 9)] # get randam word from the list
    attempt = 1    # the numebr of the user's attempts
    currentGuess = "_" * len(word)    # the process of the word
    guessedChars = []     # letter(s) that user have inputed 
    print("Welcome to Hangman!")   # print greeting 
    
    while attempt <= 10:    # when the number of attempt left is still less than 10
        guessedCharsToPrint = " ".join(guessedChars)   # print space betweet each letter
        # print out the number of attempt user have and letter that user have inputed 
        print(f"You have {11 - attempt} lives left and you have used these characters: {guessedCharsToPrint}")
        currentGuessToPrint = " ".join(currentGuess)   # print space betweet each letter
        print(f"Current word: {currentGuessToPrint}")  # the process of the word 
        
        character = input("Guess a letter: ")  # ask user to guess a letter 
        character = character.capitalize()   # Capitalize the input 
        # if user input a repeated letter, then ask user to input another letter  
        while character in guessedChars:
            print("You have already guessed this letter, try another letter.")
            print(f"You have used these characters: {guessedCharsToPrint}")
            character = input("Guess a letter: ")
            character = character.capitalize()
            
        # add the inputed letters into the list to check if user input the same letter again
        guessedChars.append(character) 
        
        newGuessWord = ""  # creat a string to update the word 
        for i in range(0, len(word)):
            if word[i] == character:
                newGuessWord = newGuessWord + character  # if the input is in the word, then update the newGuessWord
            else:
                newGuessWord = newGuessWord + currentGuess[i] # if the input is not in the word, then no update 
        
        # if there is not updates of the newGuessWord, user lost 1 attempt
        if currentGuess == newGuessWord:
            print(f"Your letter {character} is not in the word.")
            attempt += 1
        currentGuess = newGuessWord
        
        # if the current process is the same as the random picked word, then user win
        if currentGuess == word:
            print(f"You win! The word is: {word}")
            decryptText = decrypt(word)
            print(f"Decrypted word is {decryptText}")
            break
        print()
    
    # when the number of attempt is grate then 10 and the process of the word still not complete, user lost the game 
    if currentGuess != word:
        print("You lost.")
        print(f"The word is: {word}")
       
    
def decrypt(ciphertext):    # define function decrypt()
    plaintext = ""          # define plaintext
    for i in range(0, len(ciphertext)):     # iterate through the ciphertext
        if i % 2 == 0:                      # when i is even
            cipher_ascii = ord(ciphertext[i])  # convert the ciphertext character to ascii
            plain_ascii = cipher_ascii - 1     # subtract cipher_ascii by 1 to get a plain_ascii
        else:                               # when i is odd
            cipher_ascii = ord(ciphertext[i])  
            plain_ascii = cipher_ascii + 1     # add cipher_ascii by 1 to get a plain_ascii
        plaintext += chr(plain_ascii)          # append the new plain character to plaintext
    
    return plaintext
```
